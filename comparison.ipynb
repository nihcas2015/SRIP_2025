```python
import matplotlib.pyplot as plt
import numpy as np

# Define scenarios
scenarios = ['Old Model', 'RFE+Mutual Information',  'No Feature Selection',"Proposed Model"]

# Accuracy values for each target variable (converted to percentages)
attack_accuracy = [0.999897779080148, 0.9998705201681874, 0.9998705201681874,0.9998705201681874]
category_accuracy = [0.9995665832998276, 0.9999332156656967,  0.9999522969040691,0.999948208067275]
subcategory_accuracy = [0.9987283717570413, 0.9999277638833046,  0.9998309947458447,0.9999427562848829]

# Convert to percentage
attack_accuracy = [x * 100 for x in attack_accuracy]
category_accuracy = [x * 100 for x in category_accuracy]
subcategory_accuracy = [x * 100 for x in subcategory_accuracy]

# Bar width and position settings
bar_width = 0.25
x = np.arange(len(scenarios))

# Plotting
plt.figure(figsize=(14, 6))

# Plot bars
bars1 = plt.bar(x - bar_width, attack_accuracy, width=bar_width, label='Attack', color='skyblue')
bars2 = plt.bar(x, category_accuracy, width=bar_width, label='Category', color='lightgreen')
bars3 = plt.bar(x + bar_width, subcategory_accuracy, width=bar_width, label='Subcategory', color='salmon')

# Add labels above bars
def add_labels(bars):
    for bar in bars:
        height = bar.get_height()
        if height != 0:
            plt.text(bar.get_x() + bar.get_width() / 2, height + 0.001, f'{height:.4f}%', 
                     ha='center', va='bottom', fontsize=8)

add_labels(bars1)
add_labels(bars2)
add_labels(bars3)

# Labels and formatting
plt.xlabel('Scenarios')
plt.ylabel('Accuracy (%)')
plt.title('Zoomed Accuracy Comparison Across Scenarios for Attack Variables')
plt.xticks(x, scenarios, rotation=15)
plt.ylim(99.85, 100.0)  # Zoom in on y-axis to show small accuracy differences
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.tight_layout()
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Accuracy data
df_acc = pd.DataFrame({
    "Model": ["Random Forest", "XGBoost", "Logistic Regression", "Stacked Ensemble"] * 3,
    "Target": ["Attack", "Category", "Subcategory"] * 4,
    "Score": [
        0.9999, 0.9440, 0.9930,       # Random Forest
        0.9999, 0.9999, 0.9983,       # XGBoost
        1.0000, 0.9554, 0.9953,       # Logistic Regression
        0.99987, 0.99995, 0.99994     # Stacked Ensemble
    ]
})
df_acc["Metric"] = "Accuracy"

# Weighted F1 Score data
df_f1 = pd.DataFrame({
    "Model": ["Random Forest", "XGBoost", "Logistic Regression", "Stacked Ensemble"] * 3,
    "Target": ["Attack", "Category", "Subcategory"] * 4,
    "Score": [
        0.9999, 0.9440, 0.9903,       # Random Forest
        0.9999, 0.9999, 0.9982,       # XGBoost
        1.0000, 0.9554, 0.9947,       # Logistic Regression
        0.99993, 0.99997, 0.99996     # Stacked Ensemble
    ]
})
df_f1["Metric"] = "Weighted F1 Score"

# Combine both
df_all = pd.concat([df_acc, df_f1])

# Plot using Plotly
fig = px.bar(
    df_all,
    x="Target",
    y="Score",
    color="Model",
    barmode="group",
    facet_row="Metric",
    text_auto=".5f",
    title="Model Performance Comparison by Target Variable"
)

fig.update_layout(
    yaxis=dict(range=[0.94, 1.001]),
    font=dict(family="Serif", size=14),
    title_font=dict(size=20),
    legend_title_text="Model",
    bargap=0.15,
    height=800
)

fig.update_traces(textfont_size=12)
fig.show()

```
